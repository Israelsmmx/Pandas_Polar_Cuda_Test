```python
import warnings
warnings.filterwarnings('ignore')

import timeit
setup = '''
import pandas as pd
import cudf
import polars as pl
data_pl = pl.read_csv('SAPD58_SYS_TABLE_COLUMNS.csv', infer_schema_length = 1000, ignore_errors = True)
data_pd = pd.read_csv('SAPD58_SYS_TABLE_COLUMNS.csv',low_memory=False)
cuda_df = cudf.from_pandas(data_pd)
'''
#cuda_df = cudf.read_csv('SAPD58_TBTCO.csv')

data_pd_time = timeit.timeit('data_pd.groupby("TABLE_OID").sum()', setup=setup, number=100)
cuda_df_time = timeit.timeit('cuda_df.groupby("TABLE_OID").sum()', setup=setup, number=100)
data_pl_time = timeit.timeit('data_pl.groupby("TABLE_OID").sum()', setup=setup, number=100)

print(f'from Pandas')
print(f'Pandas time: {data_pd_time}')
print(f'Cudf time: {cuda_df_time}')
print(f'Polars time: {data_pl_time}')

setup = '''
import pandas as pd
import cudf
import polars as pl
data_pl = pl.read_csv('SAPD58_SYS_TABLE_COLUMNS.csv', infer_schema_length = 1000, ignore_errors = True)
data_pd = pd.read_csv('SAPD58_SYS_TABLE_COLUMNS.csv',low_memory=False)
cuda_df = cudf.read_csv('SAPD58_SYS_TABLE_COLUMNS.csv')
'''

data_pd_time = timeit.timeit('data_pd.groupby("TABLE_OID").sum()', setup=setup, number=100)
cuda_df_time = timeit.timeit('cuda_df.groupby("TABLE_OID").sum()', setup=setup, number=100)
data_pl_time = timeit.timeit('data_pl.groupby("TABLE_OID").sum()', setup=setup, number=100)

print(f'from files')
print(f'Pandas time: {data_pd_time}')
print(f'Cudf time: {cuda_df_time}')
print(f'Polars time: {data_pl_time}')
```
